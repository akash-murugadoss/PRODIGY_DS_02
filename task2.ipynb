**Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/train_and_test2.csv")
```

**Checking foe missing** **values**

```python
data.isnull().sum()
```

```python
for col in data.columns:
  if data[col].dtype == 'object':
    print(col, data[col].value_counts())
```

```python
sns.histplot(data["Age"])
plt.show()
```

**Explore relationships between variables**

```python
sns.jointplot(x="Parch", y="Fare", data=data)
plt.show()
```

**Analyze categorical vs numerical variables**

```python
sns.boxplot(x="Sex", y="Age", data=data)
plt.show()
```
